```python
car_dataset['Fuel_Type'].replace({'Petrol':0,'Diesel':1,"CNG":2},inplace=True)
car_dataset['Seller_Type'].replace({'Dealer':0,'Individual':1},inplace=True)
car_dataset['Transmission'].replace({'Manual':0,'Automatic':1},inplace=True)
```

## Splitting Data Into Training and Test Data


```python
car_dataset.columns
```

```python
X=car_dataset.drop(['Car_Name','Selling_Price'],axis=1)
Y=car_dataset['Selling_Price']
```

```python
X
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
x_train
```

```python
y_train
```

## Model Training

```python
# Loading Linear Regression

model=LinearRegression()
model.fit(x_train,y_train)
```

## Model Evaluation

```python
## Predict training Data

training_data=model.predict(x_train)
error_score=metrics.r2_score(y_true=y_train,y_pred=training_data)
# error_score=metrics.r2_score(training_data,y_train)
print("R Squared value",error_score*100)
```

```python
## Predict Testing Data

testing_data=model.predict(x_test)
error_score=metrics.r2_score(y_true=y_test,y_pred=testing_data)
# error_score=metrics.r2_score(training_data,y_train)
print("R Squared value",error_score*100)
```
